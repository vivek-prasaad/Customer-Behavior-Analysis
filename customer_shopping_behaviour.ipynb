```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower() #column names are change to lowercase
```

```python
df.columns
```

```python
df.columns=df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create a column age_group
labels=['young adult','adult','middle_aged','senior',]
df['age_group']=pd.qcut(df['age'],q=4,labels=labels) #qcut is it splits ages into 4 equal sized groups basd on the data distribution and assigns the labels we defined
```

```python
df[['age','age_group']].head(10)
```

```python
# create mapping dictionary
frequency_mapping = {
    'Weekly': 7,
    'Bi-Weekly': 14,
    'Fortnightly': 14,
    'Monthly': 30,
    'Quarterly': 90,
    'Every 3 Months': 90,
    'Annually': 365
}
# map values to new column
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
